## Data Preparation for Financial Metrics Dashboard

```python
from ScraFSY import YFinanceScrapper
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

## Data Collection

```python
agro=['AALI.JK','ANDI.JK','ANJT.JK','BWPT.JK','CBUT.JK','DSNG.JK','FAPA.JK','GZCO.JK','JAWA.JK','LSIP.JK','MGRO.JK','MKTR.JK','PSGO.JK','SGRO.JK','SIMP.JK','SMAR.JK','SSMS.JK','STAA.JK','TAPG.JK','TBLA.JK','TLDN.JK','UNSP.JK']
```

```python
#Function for scrape all agro commpanies 
def get_scrape(list_of_company):
    collect={}
    frames=[]
    frame2=[]
    for comp in list_of_company:
        collect[comp]=YFinanceScrapper(comp)
        collect[comp].get_alldata()
        collect[comp].important_dataframe()
        collect[comp].convert_to_csv('imp_dataframe')
        collect[comp].metric_dataframe()
        collect[comp].convert_to_csv('metric')
        frames.append(collect[comp].metric)
        frame2.append(collect[comp].imp_dataframe)
        df=pd.concat(frames)  
        df2=pd.concat(frame2)
    return df, df2

#function for save to csv
def save_to_csv(df,name_of_file):
    df.to_csv(name_of_file,index=False)
```

```python
df, df2= get_scrape(agro)
```

```python
save_to_csv(df,'important_metric.csv')
save_to_csv(df2,'key_feature.csv')
```

```python
#Function for check Nan value and duplicate data
def check_data(df):
    duplicate_data=df.duplicated().sum()
    nan_value=df.isna().sum()
    return duplicate_data, nan_value

#Function for check distribution of data by usig histogram
def show_hisplot(df):
    plt.figure(figsize=(40,20))
    factor=[]
    for i in range(len(df.columns)-2):
        if (len(df.columns)-2)%(i+1) == 0:
            x=i+1
            factor.append(x)
    if len(factor)>=3:
        if len(factor)%2==0:
            a=int((len(factor)/2)-1)
            b=int(len(factor)/2)
        elif len(factor)%2!=0:
            a=int((len(factor)+1)/2)
            b=int((len(factor)+1)/2)
    elif len(factor)<3:
        a=1
        b=2
    for i in range(len(df.columns)-2):
        plt.subplot(factor[a],factor[b],i+1)
        sns.histplot(df[df.columns[i+2]], kde=True)  
        
#Function for check distribution data by using boxplot     
def show_boxplot(df):
    plt.figure(figsize=(40,20))
    factor=[]
    for i in range(len(df.columns)-2):
        if (len(df.columns)-2)%(i+1) == 0:
            x=i+1
            factor.append(x)
    if len(factor)>=3:
        if len(factor)%2==0:
            a=int((len(factor)/2)-1)
            b=int(len(factor)/2)
        elif len(factor)%2!=0:
            a=int((len(factor)+1)/2)
            b=int((len(factor)+1)/2)
    elif len(factor)<3:
        a=1
        b=2
    for i in range(len(df.columns)-2):
        plt.subplot(factor[a],factor[b],i+1)
        sns.boxplot(df[df.columns[i+2]]) 

#Function for check outlier data by using IQR method        
def find_outliers_IQR(df,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    outliers = df[column][((df[column]<(q1-1.5*iqr)) | (df[column]>(q3+1.5*iqr)))]
    outlier_df=df.loc[outliers.index][['time','company',column]]
    return outliers, outlier_df

#Function for drop outlier
def drop_outlier(df, drop_list):
    df=df.drop(drop_list)
    return df
```
